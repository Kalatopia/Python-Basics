```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math

plt.figure(figsize=(27, 9))
plt.subplot(1, 3, 1)
x0 = 5
y0 = 0
r = 5
x = []
x1 = []
y = []
y1 = []

for i in range(501):
    k = i/50
    x.append(k)
    x1.append(-k)
    y.append(math.sqrt(pow(r, 2) - pow((k - x0), 2)) + y0)
    y1.append(-(math.sqrt(pow(r, 2) - pow((k - x0), 2)) + y0))
plt.plot(x, y, "r")
plt.plot(x, y1, "r")
plt.axis('equal')
plt.title("Окружность")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.subplot(1, 3, 2)
a = 5
b = 4
x = []
x1 = []
y = []
y1 = []

for i in range(501):
    k = i/100
    x.append(k)
    x1.append(-k)
    y.append(math.sqrt((1 - (pow(k, 2)/pow(a, 2)))*pow(b, 2)))
    y1.append(-math.sqrt((1 - (pow(k, 2)/pow(a, 2)))*pow(b, 2)))
plt.plot(x, y, "b")
plt.plot(x, y1, "b")
plt.plot(x1, y, "b")
plt.plot(x1, y1, "b")
plt.axis('equal')
plt.title("Эллипс")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.subplot(1, 3, 3)
a = 3
b = 7
x = []
x1 = []
y = []
y1 = []

for i in range(10):
    k = i/10 + a
    x.append(k)
    x1.append(-k)
    y.append(math.sqrt(pow(b, 2) * (pow(k, 2)/pow(a, 2) - 1)))
    y1.append(-(math.sqrt(pow(b, 2) * (pow(k, 2)/pow(a, 2) - 1))))
plt.plot(x, y, "g")
plt.plot(x, y1, "g")
plt.plot(x1, y, "g")
plt.plot(x1, y1, "g")
plt.axis('equal')
plt.title("Гипербола")
plt.xlabel('x', size=12)
plt.ylabel('y', size=12)
plt.grid()
```
