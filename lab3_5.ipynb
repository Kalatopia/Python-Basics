```python
from matplotlib import figure
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pylab import *
import numpy as np

fig = figure()
ax = Axes3D(fig)
x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x,y = np.meshgrid(x,y)
z = 2*x + 3*y
z1 = 2*(x+10) + 3*(y+10)
ax.plot_wireframe(x, y, z)
ax.plot_wireframe(x, y, z1)

```

```python
from matplotlib import figure
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pylab import *
import numpy as np

fig = figure()
ax = Axes3D(fig)
x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x,y = np.meshgrid(x,y)
z = 0.2*x**2 + 0.1*y**2 
z1 = np.sqrt(x**2 + y**2)
ax.plot_surface(x, y, z1)
ax.plot_wireframe(x-10, y-10, z)


```
