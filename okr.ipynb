```python
import matplotlib.pyplot as plt
import numpy as np
import math

a = 5
b = 0
r = 5
x = []
x1 = []
y = []
y1 = []

for i in range(101):
    k = i/10
    x.append(k)
    x1.append(-k)
    y.append(math.sqrt(pow(r, 2) - pow((k - a), 2)) + b)
    y1.append(-(math.sqrt(pow(r, 2) - pow((k - a), 2)) + b))
plt.plot(x, y)
plt.plot(x, y1)
plt.axis('equal')
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
